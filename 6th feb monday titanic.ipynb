# single group single column

### count of passengers who survived from each class

```python
Pclass_Survived = df.groupby(['Pclass'])['Survived']
```

```python
Pclass_Survived.sum()
```

# multiple columns

### total amount paid by each class who survived

```python
Pclass_Survived_fare=df.groupby(['Pclass'])['Survived','Fare']
```

```python
Pclass_Survived_fare.sum()
```

# multiple groups

#### count of females and males who survived in each class

```python
Pclass_sex_suvived=df.groupby(['Pclass','Sex'])['Survived']
```

```python
Pclass_sex_suvived.sum()
```

```python
Pclass_sex_suvived=df.groupby(['Pclass','Sex'],as_index=False)['Survived']
```

```python
Pclass_sex_suvived.sum()
```

# multiple functions

#### mean and sum of pclass &survived

```python
Pclass_Survived_mean_sum=df.groupby(['Pclass'])['Survived'].agg(['mean','sum'])
```

```python
Pclass_Survived_mean_sum
```

#### count of people survived in pclass

#### mean and sum of pclass,survived, age &fare

```python
Pclass_Survived_Age_Fare_mean_sum
```
