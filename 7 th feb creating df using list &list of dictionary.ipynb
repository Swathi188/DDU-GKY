# creating a dataframe using list

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
name=['Arun','Varun','Leela','Geetha','Salim','John']
age=[27,26,24,27,26,28]
```

```python
df=pd.DataFrame(list(zip(name,age)))
df
```

### specify the column name

```python
df=pd.DataFrame(list(zip(name,age)),columns=['Name','Age'])
df
```

## ##creating a dataframe froma list of dictionaries

```python
list_of_dictionary=[
    {'Name':'Varun','Age':27},
    {'Name':'Arun','Age':26},
    {'Name':'Leela','Age':24},
    {'Name':'Geetha','Age':27},
    {'Name':'Salim','Age':26},
    {'Name':'John','Age':28}
]
```

```python
df_dict=pd.DataFrame(list_of_dictionary)
df_dict
```

### creating a new dataframe using list method

```python
name=['Rita','Arun','Sita','Varun','Ram','Radha','Mohan','Devi','Nidhi']
age=[23,29,14,21,32,23,51,20,29]
dob=['20/02/97','01/07/91','07/07/06','01/05/99','07/11/88','06/09/97','03/03/69','05/01/00','10/01/91']
gender=['f','m','f','m','m','f','m','f','f']
```

```python
df=pd.DataFrame(list(zip(name,age,dob,gender)),columns=['name','age','dob','gender'])
df
```

## forward fill (row)


```python
df.fillna(method='ffill')
```

## backward fill(row)

```python
df.fillna(method='bfill')
```

# limiting the forward /backward fill

```python
df.fillna(method='ffill',limit=1)
```

### filling with pandas objects

```python
df.fillna(df.mean())
```
