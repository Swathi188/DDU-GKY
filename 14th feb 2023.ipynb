## PLOTING WITH PANDAS

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('Car_sales.csv')
df.sample(10)
```

#### droping na values

```python
df.dropna(inplace=True)
```

```python
df.info()
```

### plots on numerical value columns

### Histogram

```python
df.Horsepower.plot.hist(figsize=(10,10),cmap='brg');
```

#### Box plot

```python
df.Sales_in_thousands.plot.box(figsize=(4,3), cmap='gist_earth');
```

```python
df.Horsepower.plot.box(figsize=(4,3), cmap= 'flag');
```

#### subplots

```python
df.plot.box(figsize=(22,4),subplots=True);
```

```python
df.plot.hist(figsize=(6,18),subplots=True);
```

### LINE CHARTS

```python
df.sort_values(by='__year_resale_value').plot.line('__year_resale_value','Price_in_thousands',figsize=(6,3), cmap='inferno');
```

#### BAR PLOT

```python
df.Manufacturer.value_counts().plot.bar(figsize=(6,2), cmap='twilight_shifted');
```

```python
df.sort_values(by='Price_in_thousands')[:20].plot.bar('Manufacturer','Price_in_thousands',figsize=(6,3));
```

#### AREA CHART

```python
df.plot.area('Manufacturer','Price_in_thousands', figsize=(6,3), cmap='Pastel2');
```

```python
df.plot.area('Manufacturer','Price_in_thousands', figsize=(6,3), cmap='ocean');
```

```python
df.plot.area('Manufacturer','Price_in_thousands', figsize=(6,3), cmap='CMRmap');
```

#### Horizontal bar chart

```python
df.sort_values(by='Price_in_thousands')[:5].plot.barh('Model','Power_perf_factor',figsize=(6,4));
```

```python
df.sort_values(by='Price_in_thousands')[:5].plot.barh('Model','Power_perf_factor',figsize=(6,4),cmap='Purples_r');
```

### Scatter Plot

```python
df.sort_values(by='Price_in_thousands')[:5].plot.scatter('Latest_Launch','Price_in_thousands' , figsize=(5,2));
```

```python
df.plot.box(figsize=(22,5));
```

```python
df.sort_values(by='Engine_size',ascending=False)[:10].Engine_size.plot.pie();
```

```python
df.sort_values(by='Engine_size',ascending=False)[:3].Engine_size.plot.pie();
```

### KDE Plot

```python
df.plot.kde(figsize=(10,5));
```

### subplots

```python
df.plot.kde(figsize=(12,10),subplots=True);
```

### Hexbin plot

```python
df.plot.hexbin('Curb_weight','Fuel_capacity',gridsize=15, cmap='autumn');
```

```python
df.plot.hexbin('Curb_weight','Fuel_capacity',gridsize=15, cmap='Set3');
```

```python
df.plot.hexbin('Curb_weight','Fuel_capacity',gridsize=15, cmap='gnuplot');
```
