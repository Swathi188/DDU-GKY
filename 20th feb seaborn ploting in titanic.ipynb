```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df=pd.read_csv('titanic.csv')
df.head()
```

#### UNIVARIATE ANALYSIS

#### KDE PLOT

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=df.PassengerId)
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=df.Age)
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.kdeplot(data=df.Fare)
plt.show()
```

### HISTPLOT

```python
sns.histplot(df.Fare)
plt.show
```

#### BOX PLOT

```python
sns.boxplot(df.Age)
plt.show()
```

#### MULTI VARIATE ANALYSIS

#### LINE PLOT

```python
sns.lineplot('Age','Fare',data=df)
plt.title('Age vs Fare')
plt.show()
```

```python
sns.lineplot(x='Age', y='Fare', data=df)
plt.title('Age vs Fare')
plt.show()
```

```python
sns.lineplot('Age','Fare',data=df, hue='Sex')
plt.title('Age vs Fare')
plt.show()
```

```python
sns.lineplot('Age','Fare',data=df, hue='Pclass')
plt.title('Age vs Fare')
plt.show()
```

```python
sns.lineplot('Age','Fare',data=df, hue='Survived')
plt.title('Age vs Fare')
plt.show()
```

### PIE CHART 

```python
df.columns
```

```python
pclass_survived=df.groupby(['Pclass'])['Survived'].sum()
```

```python
pclass_survived
```

```python
sns.set_style('ticks')
pclass_survived.plot.pie()
plt.legend()
plt.show()
```

```python
pclass_sex_Survived=df.groupby(['Pclass','Sex'])['Survived'].sum()
```

```python
pclass_sex_Survived
```

```python
pclass_sex_Survived.plot.pie(autopct = '%1.2f%%')
plt.legend(bbox_to_anchor=(1.5,1),loc='upper left',borderaxespad=0)
plt.show()
```

#### BAR CHART

```python
plt.figure(figsize=(6,4))
sns.countplot('Sex',hue='Survived',data=df)
plt.show()
```

### COUNT PLOT

```python
plt.figure(figsize=(4,3))
sns.countplot('Survived',data=df)
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.set_style('darkgrid')
sns.countplot('Pclass',hue='Survived',data=df)
plt.title('Pclass:Survived vs Dead')
plt.show()
```

```python
plt.figure(figsize=(4,3))
sns.set_style('darkgrid')
sns.countplot('Pclass',hue='Sex',data=df)
plt.title('Pclass:Sex vs Dead')
plt.show()
```

### FACTOR  PLOT

```python
sns.set_style('ticks')
sns.factorplot('Pclass','Survived',hue='Sex',data=df)
plt.show()
```

### violin plot

```python
sns.set_style('dark')
plt.figure(figsize=(4,3.5))
sns.violinplot('Sex','Age',hue='Survived',data=df)
plt.show()
```

#### BAR PLOT

```python
plt.figure(figsize=(8,4))
sns.barplot('SibSp','Survived',data=df)
plt.title('SibSp & Survived')
plt.show()
```

### Pair plot

```python
sns.pairplot(data=df)
plt.show()
```

### Heat map

```python
heat_map=df.corr()
sns.heatmap(heat_map)
plt.show()
```
