### Extracting year month day from the date column

```python
df['year']=pd.DatetimeIndex(df.Date).year
df['month']=pd.DatetimeIndex(df.Date).month
df['day']=pd.DatetimeIndex(df.Date).day
```

```python
df.head()
```

### drop lat and long

```python
df.drop(columns=['Lat','Long','Date'],inplace=True)
```

```python
df
```

### converting month number to name

```python
df.month.replace({
    1:'january',
    2:'february',
    3:'march',
    4:'april',
    5:'may',
    6:'june',
    7:'july',
    8:'august',
    9:'september',
    10:'october',
    11:'november',
    12:'december'
},inplace=True)
```

```python
df
```

### Rename column Country/Region to Region

```python
df.rename(columns={'Country/Region':'Region'},inplace=True)
```

```python
df.rename(columns={'WHO Region':'WHO_Region'},inplace=True)
```
